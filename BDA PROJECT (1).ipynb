# Feature engineering and Linear Regression

So as we knew that the income, spending score, and credit score is affecting the Annual Spending so we include only them other than that will be dropped and we will encode prefferd category into numeric form as ML does not understand categorical data it understands numeric values

```python
data.head()
```

```python
data.AnnualIncome.value_counts()
```

```python
data.drop(['CustomerID','Age','TenureYears','LoyaltyPoints','Gender'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.shape
```

```python
# let's start tranaing on the dataset 
from sklearn.model_selection import  train_test_split
from sklearn.linear_model import  LinearRegression
```

```python
data.columns
```

```python
X = data[['AnnualIncome', 'SpendingScore', 'CreditScore']]
y = data[['AnnualSpending']]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=43)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
# let's evalutate whether our y_predicted is how much near the actual y value
from sklearn.metrics import  mean_squared_error,r2_score
```

```python
MSE = mean_squared_error(y_test,y_predicted)
print(f"MSE: {MSE}")
print(f"RMSE: {np.sqrt(MSE)}")
```

```python
performance=r2_score(y_test,y_predicted)
print(f"r_2 SCORE: {performance}")
```

```python
# let's plot this to check how well is this working 
plt.figure(figsize=(8, 5))

# Plotting each feature in X_test against y_test and y_predicted
for i, feature in enumerate(X_test.columns):
	plt.scatter(X_test[feature], y_test, color='blue', label='Actual Annual Spending', alpha=0.7)
	plt.scatter(X_test[feature], y_predicted, color='red', label='Predicted Annual Spending', alpha=0.7)
	plt.plot(X_test[feature], y_predicted, color='green', label='Regression Line')
	plt.title(f'Actual vs Predicted Annual Spending for {feature}')
	plt.xlabel(feature)
	plt.ylabel('AnnualSpending')
	plt.legend()
	plt.show()
```
